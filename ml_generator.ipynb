```python
import numpy as np
```

```python
def generate_array(start, stop):
    values = np.linspace(start, stop, 10)
    variance = 0.1 * values
    values += np.round(np.random.uniform(low=-variance, high=variance, size=10),3)
    print("[",", ".join(str(x) for x in values),"]",end="")
```

```python
array = [[0,100,0,200,0,300,0,1000],[0,1000,0,200,0,5,0,100],[16,32,0,20,0,100,-10,200]]
for i in range (3):
    x = array[i]

    print("X = np.array([")
    generate_array(x[0],x[1])
    print(",")
    generate_array(x[2],x[3])
    print(",")
    generate_array(x[4],x[5])
    print("\n])")
    print("X = np.transpose(X)")
    print("y = np.array(", end="")
    generate_array(x[6],x[7])
    print(")")
    print("\n")
```

```python
x = [0,1000,0,100,-10,200,0,100]

print("X = np.array([")
generate_array(x[0],x[1])
print(",")
generate_array(x[2],x[3])
print(",")
generate_array(x[4],x[5])
print("\n])")
print("X = np.transpose(X)")
print("y = np.array(", end="")
generate_array(x[6],x[7])
print(")")
print("\n")
```
